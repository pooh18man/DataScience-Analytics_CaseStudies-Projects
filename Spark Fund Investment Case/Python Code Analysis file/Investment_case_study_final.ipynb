## Part 5.1: For 'USA'

```python
## 'D1' is the dataframe for country 'USA'

# necessary to fillna values in dataframe by '0' in the 'Sector_frame'
Sector_frame['raised_amount_usd'].fillna(0, inplace=True)
D1 = Sector_frame.loc[Sector_frame['country_code'] == 'USA']

# Range of investment made by Spark Funds i.e., between 5000000 and 15000000 millions USD
D1 = D1[D1['raised_amount_usd'].between(5000000, 15000000, inclusive=True)]

# Number of investments in each of the 'main_sector'
D1['invest_count'] = D1['Main_sector'].groupby(D1['Main_sector']).transform('count')

# Amount of investment made in each 'main_sector'
D1['sum_invest'] = D1.groupby('Main_sector')['raised_amount_usd'].transform('sum')
D1.shape
```

```python
# Total number of investments made for 'USA'

total_investments_USA = D1.groupby(['Main_sector', 'invest_count'], as_index=False).count()
total_investments_USA['invest_count'].sum()
```

```python
# Total amount of investments made for 'USA'

total_sum_USA = D1.groupby(['Main_sector','sum_invest'], as_index=False).count()
total_sum_USA['sum_invest'].sum()
```

```python
# find the main sector with max number of investments
total_investments_USA.groupby('Main_sector')['invest_count'].sum().sort_values(ascending=False)
```

```python
## Following code is written for exporting data to csv file and plotting the same on Tableau
us = total_investments_USA.groupby('Main_sector')['invest_count'].sum().sort_values(ascending=False)
us = pd.DataFrame(us)
us
```
